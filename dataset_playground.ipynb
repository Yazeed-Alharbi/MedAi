```python
medical_conditions = [
    'Hypertension', 'Diabetes', 'Asthma', 'Heart Disease', 'Arthritis',
    'Depression', 'Allergies', 'Migraine', 'Cancer', 'Chronic Pain'
]

medications = [
    'Metformin', 'Lisinopril', 'Atorvastatin', 'Albuterol', 'Simvastatin',
    'Omeprazole', 'Levothyroxine', 'Amlodipine', 'Hydrochlorothiazide', 'Gabapentin'
]

allergies = [
    'Pollen', 'Dust Mites', 'Pet Dander', 'Mold', 'Peanuts',
    'Shellfish', 'Latex', 'Penicillin', 'Bee Stings', 'Soy'
]
```

```python

np.random.seed(42)  

# Randomly assign new values from the lists
df['medical_conditions'] = np.random.choice(medical_conditions, size=len(df))
df['medications'] = np.random.choice(medications, size=len(df))
df['allergies'] = np.random.choice(allergies, size=len(df))

```

```python
df
```

```python
import pandas as pd
from datetime import datetime

df['date_of_birth'] = pd.to_datetime(df['date_of_birth'])

# Calculate age as the difference between today's date and the date_of_birth
df['age'] = df['date_of_birth'].apply(lambda x: (datetime.now() - x).days // 365)

# Rename the 'date_of_birth' column to 'age'
df.rename(columns={'date_of_birth': 'age'}, inplace=True)
```

```python
df = df.loc[:, ~df.columns.duplicated(keep='last')]
```

```python
np.random.seed(42)  # For reproducible results

# Randomly select three distinct medical conditions, medications, and allergies for each row
df['medical_conditions'] = [np.random.choice(medical_conditions, 3, replace=False).tolist() for _ in range(len(df))]
df['medications'] = [np.random.choice(medications, 3, replace=False).tolist() for _ in range(len(df))]
df['allergies'] = [np.random.choice(allergies, 3, replace=False).tolist() for _ in range(len(df))]
```
