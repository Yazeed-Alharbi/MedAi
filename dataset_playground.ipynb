```python
import pandas as pd
import numpy as np
from datetime import datetime
import random

appointment_titles = [
    'MRI', 'CT Scan', 'Ultrasound', 'Consultation', 'Routine Check-up',
    'Surgery', 'Therapy Session', 'Dental Check', 'Vaccination', 'Emergency Care'
]

# Cities and institutions within the Kingdom of Saudi Arabia (KSA)
cities = ['Riyadh', 'Jeddah', 'Mecca', 'Medina', 'Dammam']
city_institutions = {
    'Riyadh': ['King Faisal Specialist Hospital', 'King Saud University Medical City'],
    'Jeddah': ['King Abdulaziz Medical City'],
    'Mecca': ['Al Noor Specialist Hospital'],
    'Medina': ['King Fahad Hospital'],
    'Dammam': ['Dhahran Health Center', 'Dammam Central Hospital']
}

# Lorem Ipsum text for the conclusion
conclusion_text = "Lorem ipsum dolor sit amet, consectetur adipiscing elit."

# Function to generate random dates within a range
def random_dates(start, end, n):
    start_u = start.value // 10**9
    end_u = end.value // 10**9
    return pd.to_datetime(np.random.randint(start_u, end_u, n), unit='s')

# Start and end dates for the random date generation (e.g., between 2020 and 2023)
start = pd.to_datetime('2020-01-01')
end = pd.to_datetime('2023-12-31')


n = len(df) 

# Add columns with randomized data
df['appointment_title'] = np.random.choice(appointment_titles, size=n)
df['appointment_city'] = np.random.choice(list(city_institutions.keys()), size=n)
df['appointment_date'] = random_dates(start, end, n)

# Ensure that the institution matches the city
df['appointment_institution'] = df['appointment_city'].apply(lambda x: np.random.choice(city_institutions[x]))


df['appointment_conclusion'] = conclusion_text

```
